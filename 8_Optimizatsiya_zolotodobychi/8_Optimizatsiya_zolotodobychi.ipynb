### Целевой признак. Проверка подсчёта

Прежде, чем приступить к "причёсыванию" пропусков и пр., нужно проверить главное!
Целевой признак у нас в проекте - эффективность обогащения. В данных есть два столбца с этими признаками: для чернового концентрата и для финального. Проверим на примере чернового в тестовой выборке, правильно ли рассчитана эффективность: вычислим эффективность сами и сравним метрикой MAE. 

Если окажется, что наши с исходными расчёты сильно различаются, мы зря тут вообще все собрались!

Итак, из миллиона имеющихся признаков надо найти:
1. C - доля золота после флотации. Наиболее вероятно, что это rougher.output.concentrate_au
2. F - доля золота в сырье до флотации. Максимально похож на это признак rougher.input.feed.au
3. T - доля золота в отвальных хвостах после флотации. Это без всяких сомнений rougher.output.tail_au

Напишем функцию подсчёта и применим построчно к тестовой выборке.

```python
# выделим нужные столбцы и удалим для подсчёта строки, где параметр F равен нулю, чтобы не произошло деления на него:
train_x = (train.loc[:, ['rougher.output.concentrate_au',
                   'rougher.input.feed_au',
                   'rougher.output.tail_au',
                   'rougher.output.recovery']]
                 .drop(train[train['rougher.input.feed_au']==0].index))

# функция для подсчёта эффективности обогащения:
def recovery(C, F, T):
    recovery = (C*(F-T)*100)/(F*(C-T))
    return recovery


# применение функции построчно к обучающей выборке (точнее, её временной копии)
train_x['recovery_check'] = train_x.apply(lambda row: recovery(row['rougher.output.concentrate_au'], 
                                             row['rougher.input.feed_au'], 
                                               row['rougher.output.tail_au']), axis=1)
train_x.head()
```

Выглядит потрясающе, как будто всё было правильно посчитано! Но все 17к строк глазами проверять не хочется. Посчитаем среднюю абсолютную ошибку, чтобы убедиться окончательно и бесповоротно, предварительно удалив строки, где пропущен исходный целевой признак.

```python
# удалим пропуски целевого признака, чтобы питон не ругался:

train_x = train_x.dropna(subset=['rougher.output.recovery'], axis=0)
print("Пропусков в целевом признаке после обработки:", train_x['rougher.output.recovery'].isna().sum())
```

```python
print("Средняя абсолютная ошибка между исходными значениями эффективности и посчитанными вручную:", 
      sk.metrics.mean_absolute_error(train_x['recovery_check'], train_x['rougher.output.recovery']))
```

Довольно скромное значение! Это хороший знак, подсчёты верны, работаем дальше.

### Предобработка

Как говорит муж, будем есть слоника по частям. Сначала сравним признаки в трёх сетах.

#### Сравнение признаков в выборках

```python
print("Размер обучающей выборки:", train.shape)
print("Размер тестовой выборки:", test.shape)
print("Размер всей выборки:", full.shape)
```

В тестовой выборке чего-то не хватает! Чего же?

```python
diff = full.columns.difference(test.columns).tolist()
print("Признаки, которые не попали в тестовую выборку:")
print(pd.DataFrame(diff))
```

В тестовую выборку по каким-то причинам не попали все параметры, касающиеся продукта (то, что на выходе после каждой из стадий), а также все параметры, которые вычислялись на основе "сырых" показателей. 

Так как тестовая выборка по условию задачи - это то, что будет доступно машине для предсказаний, тренировать модель на бОльшем количестве признаков нечестно, структурно тестовая и обучающая выборки должны быть идентичны для точной работы машины.

Поэтому уберём из обучающей выборки все признаки (кроме целевых), не вошедшие в тестовую.
А также добавим целевые признаки в тестовую выборку, иначе будет сложновато оценивать качество предсказаний на ней ;)

```python
# убираем из списка признаков два целевых, они нам понадобятся в обучающей выборке:
diff = [x for x in diff if x not in ('rougher.output.recovery', 'final.output.recovery')]

# чистим обучающую выборку от признаков, которых нет в тестовой (кроме целевых):
train = train.drop(diff, axis=1)

# добавляем целевые признаки в тестовую выборку:
test = pd.merge(test, full[['rougher.output.recovery', 'final.output.recovery', 'date']], 
                  on='date',
                  how='left')

print("Размер обучающей выборки после удаления неиспользуемых признаков:", train.shape)
print("Размер обучающей выборки после удаления неиспользуемых признаков:", test.shape)
```

Отлично. Количество столбцов одинаковое у обучающей и тестовой выборок, количество строк не изменилось. 

Теперь займёмся пропусками в самих строках. Для начала убедимся, что из full таблицы всё перенеслось в train и test.

#### Пропуски в признаках

```python
# убираем из полного сета лишние колонки, но старую версию оставим на всякий, поэтому в новую переменную:
full_x = full.drop(diff, axis=1)

# считаем количество пропусков в каждом столбце каждого сета, объединяем результаты в одну таблицу,
na_all = (pd.merge(
    pd.merge(pd.DataFrame(full_x.isna().sum()), 
             pd.DataFrame(train.isna().sum()),
             on=full_x.isna().sum().index, how='inner', 
             left_index=True, right_index=True), 
    pd.DataFrame(test.isna().sum()), 
    on=test.isna().sum().index, 
    left_index=True, right_index=True)
.reset_index(drop=True))
na_all.columns = ["parameter", "full_na_count", "train_na_count", "test_na_count"]
# добавляем столбец с суммой пропусков в тестовой и обучающей выборках
na_all["sum_train_test_na_count"] = na_all["train_na_count"]+na_all["test_na_count"]
# добавляем столбец с проверкой: отличается ли количество пропусков в полной выборке от 
# суммы пропусков в обучающей и тестовой выборках
na_all["missing_values_from_full"] = (na_all['full_na_count'] != na_all['full_na_count'])
na_all

full_x.sample(10)
```

Немного массивное исследование пропусков получилось, ну и ладно. Мы теперь знаем, что всё, что было, перенесено в train и test без потерь.

В добавление к этому, проверим, нет ли пересечений у train и test:

```python
pd.merge(train, test, how='inner', on=['date'])
```

Отлично, пересечений нет. Разделили честно!

Теперь посмотрим на пропуски в строках.

У нас имеются пропуски практически во всех столбцах, что, согласно комментариям "заказчика", вполне нормально - потому что некоторые данные могли быть ещё не готовы на момент их считывания. Так как по сути мы таки имеем дело с данными, протяжёнными во времени с разницей в один час, логично предположить, что параметры в соседние часы могуть быть весьма схожи (и да, в условии есть намёк). Это предположение можно использовать для выбора стратегии заполнения пропусков в "сырых" параметрах: заполним их соседними реальными значениями. Желательно для бОльшей точности заменить в полной выборке (больше вероятность проставить более близкое по времени значение), и затем переместить полученные значения в соответствующие объекты обучающей и тестовой выборок. Точнее, мы просто перепишем заново эти две выборки. 

Итак, план такой:
- Берём полную выборку (можно уже ту, где только нужные параметры), заменяем везде, кроме целевых, пропуски на ближайшие "живые" значения
- Добавляем значения параметров в обучающей и тестовой выборках, ориентируясь на даты
    

# АЛЛО!!!! ТУТ НАДО ПЕРЕНЕСТИ БЫВШИЕ ПУСТЫЕ ЗНАЧЕНИЯ ИЗ ФУЛЛ_Х В ОБУЧАЮУЩУЮ И ТЕСТОВУЮ!!!! ПО ДАТЕ. 

```python
# удалим целевые признаки из полной таблицы (её варианта без лишних признаков)
# заполняем везде пропуски методом ffill, где на пустое место встаёт ближайшее "живое" значение в столбце
# задаём индекс датасету в виде колонки с датами, чтобы удобнее было переносить данные
full_x = (full_x.drop(['rougher.output.recovery', 'final.output.recovery'], axis=1)
            .fillna(method='ffill')
            .set_index('date'))

# аналогично "обзываем" индексы тестовой и обучающей выборок датами из столбца date
train = train.set_index('date')
test = test.set_index('date')

# заполняем пропуски в обучающей и тестовой выборках значениями из full(_x) датасета
train = train.fillna(full_x)
test = test.fillna(full_x)
```

#### Пропуски в целевых признаках. Разделение выборок по стадиям

Что касается целевого признака recovery, то здесь лучше поостеречься и удалить пропуски, чтобы не обманывать нашу машинку бутафорскими значениями, иначе потом она начнёт обманывать (непреднамеренно, конечно!) заказчика.

Пропуски есть в обоих целевых признаках. Нужно посмотреть, совпадают ли координаты и количества этих пропусков.

```python
# сравниваем индексы строк, где расположены пропуски для двух целевых признаков, и выводим индексы несовпадения пропусков:
print("Всего пропусков в целевом признаке эффективности финальной разработки:",
      len(full[full['final.output.recovery'].isna()].index))

print("Из них", len(
full[full['final.output.recovery'].isna()].index
    .difference(full[full['rougher.output.recovery'].isna()].index)),
     "объектов имеют значение целевого признака эффективности черновой разработки.")

print("Всего пропусков в целевом признаке эффективности черновой разработки:",
    len(full[full['rougher.output.recovery'].isna()].index))

print("Из них", len(
full[full['rougher.output.recovery'].isna()].index
    .difference(full[full['final.output.recovery'].isna()].index)),
      "объектов имеют значение целевого признака эффективности финальной разработки.")
```

Видим, что пропуски у целевых признаков местами в разных местах). Значит, для экономии данных нужно уже сейчас их поделить, а затем в каждой подвыборке удалить строки с пропусками в соответствующем целевом признаке. 

Очень важно поделить данные логично! Например, признаки, касающиеся первой и второй очисток, никак не могут влиять на эффективность разработки на стадии чернового концентрата, ведь их, хмм, как бы ещё не существует)

А вот на финальную эффективность может влиять всё, потому что после уже ничего не происходит.

Итого, на этом этапе план такой:
    - Делим тестовую и обучающую выборки на "черновую" и "финальную" части
    - Удаляем такие строки в новых обучающих и тестовых выборках, в которых имеются пропуски значений соответствующего целевого признака


```python
# ищем столбцы со словом rougher и отбираем их в черновые тестовую и обучающую выборки из соответствующих сетов
rougher_train = train.loc[:, [col for col in train.columns if 'rougher' in col]]
rougher_test = test.loc[:, [col for col in train.columns if 'rougher' in col]]
# удалим для перфекционизма ненужный целевой признак из финальных выборок
train = train.loc[:, train.columns != 'rougher.output.recovery']
test = test.loc[:, test.columns != 'rougher.output.recovery']

# удаляем во всех четырёх выборках строки с пропусками в целевом признаке
train = train.dropna(axis=0)
test = test.dropna(axis=0)
rougher_train = rougher_train.dropna(axis=0)
rougher_test = rougher_test.dropna(axis=0)

print("Итоговые размеры выборок: \n", 
     "Черновая обучающая:", rougher_train.shape, "\n",
     "Черновая тестовая:", rougher_test.shape, "\n",
     "Финальная обучающая", train.shape, "\n",
     "Финальная тестовая", test.shape, "\n",
     "Соотношение выборок между собой:\n"
     "Черновые:", 
      "{0:.0%}".format(len(rougher_train)/(len(rougher_train)+len(rougher_test))), 
      "и", "{0:.0%}".format(len(rougher_test)/(len(rougher_train)+len(rougher_test))),
     "\nФинальные:", 
      "{0:.0%}".format(len(train)/(len(train)+len(test))), 
      "и", "{0:.0%}".format(len(test)/(len(train)+len(test))))
```

Фух. Техническую предобработку вроде бы завершили. Плюс пропорция между обучающей и тестовой выборками вполне достойная.

#### Корреляция признаков

Проверим признаки на коллинеарность. Возможно, не все они с нами отправятся с нами в следующий этап. Сначала построим матрицу корреляций на полном датасете (без удалённых ранее признаков). 

```python
# сначала вернём целевые признаки нашему урезанному полному датасету, корреляция с ними тоже интересна
full_x['rougher.output.recovery'] = full['rougher.output.recovery'].values
full_x['final.output.recovery'] = full['final.output.recovery'].values

# это чтобы матрица была треугольная
mask = np.triu(full_x.corr())
# строим матрицу
figure, ax = plt.subplots(figsize=(12,12))
figure = (sns.heatmap(full_x.corr(), square=True, mask=mask,
                     xticklabels=1, yticklabels=1, ax=ax, cmap='viridis_r')
 .set_title('Матрица корреляций признаков в полной выборке')
)
```

На этой великолепной тепловой карте можно увидеть следующее:
1. Отрицательной корреляции, требующей от нас каких-то действий по удалению признаков, здесь нет. Примем, что если корреляция больше 0.7, то один из признаков точно нужно удалить.
2. Присутствует довольно много высококоллинеарных в положительном смысле пар признаков. 
3. Видна кластерность расположения коллинеарных признаков. Если посмотреть на группы признаков, расположенных в этих островках, то их может объединять какой-то параметр!
- Так, видно, например, что характеристики воздуха и жидкости для флотационных установок отлично коррелируют в рамках одной стадии разработки. 
4. Но есть и отдельно стоящие пары признаков, имеющие высокую корреляцию.
5. Целевые признаки ожидаемо неплохо коррелируют друг с другом, остальные предварительно коррелируют ниже порога в 0.5.

На этот раз план выглядит так:
- Отбираем признаки с самыми высокими значениями корреляции (больше 0.7) и удаляем их из датасета
- Проверяем матрицу корреляций ещё раз
- Если всё в порядке, удаляем эти же признаки в обучающих и тестовых выборках



```python
# запишем матрицу корреляций в переменную, и возьмём корреляции по модулю (ну а вдруг на картинке не заметили строгую
# отрицательную корреляцию)
corr_matrix = full_x.corr().abs()

# цикл, который составляет список названий столбцов, в которых присутствует значение, большее 0.7
to_drop = []
for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if 1 > corr_matrix.iloc[i, j] > 0.7:
            colname = corr_matrix.columns[i]
            to_drop.append(colname)

# убираем повторяющиеся названия, которые получились вследствие работы цикла (после построчного перебора добавлялись
# названия одного и того же столбца)            
to_drop = list(set(to_drop))
full_x_dropped = full_x.drop(to_drop, axis=1)

# это чтобы матрица была треугольная
mask2 = np.triu(full_x_dropped.corr())
# строим матрицу
figure2, ax2 = plt.subplots(figsize=(12,12))
figure2 = (sns.heatmap(full_x_dropped.corr(), square=True, mask=mask2,
                     xticklabels=1, yticklabels=1, ax=ax2, cmap='viridis_r')
 .set_title('Матрица корреляций признаков в полной выборке после удаления мультиколлинеарных признаков')
)
```

Ух! Осталось не так много признаков, два из них - целевые. В выборке для чернового концентрата будет и того меньше нецелевых признаков. Что же, с этим тоже можно работать. В конце концов, если обнаружится, что машина предсказывает плоховато, сможем вернуться в этот раздел и увеличить порог допустимой корреляции.


Удалим обнаруженные мультиколлинеарности в тестовых и обучающих подвыборках и приступим к масштабированию.

```python
# Устраняем мультиколлинеарные признаки в финальных выборках
train = train.drop(to_drop, axis=1)
test = test.drop(to_drop, axis=1)

# отбираем из списка удаляемых признаков те, что относятся к черновым
to_drop_rougher = [s for s in to_drop if 'rougher' in s]

# Устраняем мультиколлинеарные признаки в черновых выборках
rougher_train = rougher_train.drop(to_drop_rougher, axis=1)
rougher_test = rougher_test.drop(to_drop_rougher, axis=1)
```

- построить распределения этих сумм
- посмотреть на них и поискать аномалии
- решить вопрос об удалении аномалий, если таковые обнаружатся, из обучающей и тестовой выборок

### Способ оценки качества моделей

Готовим сани летом. Так как нам предстоит кросс-валидационный характер обучения и проверки моделей на обучающей выборке, необходимо заранее задуматься о том, какая метрика будет оценивать качество обучения и предсказания прямо в процессе кросс-валидации. В нашем исследовании это будет sMAPE (симметричное среднее абсолютное процентное отклонение). В доументации заказчика указана формула, по которой нужно считать данную метрику, по ней и создадим специальную функцию.

В добавление ко внутренней метрике в конце работы с обучающей выборкой сравним результат с кросс-валидацией дамми-модели.

```python
# функция будет указана внутри другой предустановленной функции в качестве метрики качества
# так как все стандартные метрики качества берут на вход таргет и предсказания, 
# наша функция тоже будет их принимать на вход
def sMAPE(target, predictions):
    smape = 100/len(target) * np.sum(2 * np.abs(predictions - target) / (np.abs(target) + np.abs(predictions)))
    return smape
```

### Подбор модели на основе результатов кросс-валидации

Возьмём стандартный набор моделей регрессии: линейную регрессию, дерево и лес. Для дерева подберём глубину, для леса - глубину и количество деревьев. Для кросс-валидаций будем делить обучающую выборку на пять частей. Поехали.

#### Линейная регрессия

```python
# создаём модель линейной регрессии
model_LinearRegression = LinearRegression()
```

```python
# считаем sMAPE для обучающей черновой и обучающей финальной выборок методом кросс-валидации
sMAPE_LinearRegression = cross_val_score(model_LinearRegression, 
                                         features_train_scaled, 
                                         target_train, cv=5, 
                                         scoring=make_scorer(sMAPE, 
                                                             greater_is_better=False))
sMAPE_LinearRegression.mean()
sMAPE_rougher_LinearRegression = cross_val_score(model_LinearRegression, 
                                         features_rougher_train_scaled, 
                                         target_rougher_train, cv=5, 
                                         scoring=make_scorer(sMAPE, 
                                                             greater_is_better=False))
print("Среднее sMAPE в результате кросс-валидации на обучающей финальной выборке:", 
      abs(sMAPE_LinearRegression.mean()))
print("Среднее sMAPE в результате кросс-валидации на обучающей черновой выборке:", 
      abs(sMAPE_rougher_LinearRegression.mean()))
```

На ноль эти ошибки не очень похожи, значит, линейные регрессоры многовато ошибаются в процессе работы. Наверняка есть модели получше!

#### Дерево решений

```python
# так как у модели дерева есть гиперпараметры, будем работать сразу в цикле, перебирая их и кросс-валидируя внутри циклов
# модули метрики берутся, потому что могут быть отрицательные значения ошибки

best_model_DecisionTree = None
best_rougher_model_DecisionTree = None
best_sMAPE_DecisionTree = -100
best_sMAPE_rougher_DecisionTree = -100

#final set
for depth in range(1,21):
    
    model_DecisionTree = DecisionTreeRegressor(random_state=42, max_depth=depth)
    
    # final train set cross-validation sMAPE calculation
    sMAPE_DecisionTree = cross_val_score(model_DecisionTree, 
                                         features_train_scaled, 
                                         target_train, cv=5, 
                                         scoring=make_scorer(sMAPE, 
                                                             greater_is_better=False)).mean()
 
   
    # choice of best absolute values of sMAPE for final set
    if sMAPE_DecisionTree > best_sMAPE_DecisionTree:
        best_model_DecisionTree = model_DecisionTree
        best_sMAPE_DecisionTree = sMAPE_DecisionTree        

#rougher set        
for depth in range(1,21):
    
    model_rougher_DecisionTree = DecisionTreeRegressor(random_state=42, max_depth=depth)
        
    # rougher train set cross-validation sMAPE calculation
    sMAPE_rougher_DecisionTree = cross_val_score(model_rougher_DecisionTree, 
                                         features_rougher_train_scaled, 
                                         target_rougher_train, cv=5, 
                                         scoring=make_scorer(sMAPE, 
                                                             greater_is_better=False)).mean()
    
    # choice of best absolute values of sMAPE for rougher set
    if sMAPE_rougher_DecisionTree > best_sMAPE_rougher_DecisionTree:
        best_rougher_model_DecisionTree = model_rougher_DecisionTree
        best_sMAPE_rougher_DecisionTree = sMAPE_rougher_DecisionTree      

# чтобы получить глубину, обучаем лучшие модели, затем выводим метрику и гиперпараметр на экран       
best_model_DecisionTree.fit(features_train_scaled, target_train)
best_model_DecisionTree_depth = best_model_DecisionTree.get_depth()          
print("Среднее sMAPE в результате кросс-валидации наилучшей модели дерева решений на обучающей финальной выборке:\n", 
      best_sMAPE_DecisionTree)
print("Глубина наилучшей модели дерева:", best_model_DecisionTree_depth)   

best_rougher_model_DecisionTree.fit(features_rougher_train_scaled, target_rougher_train)
best_rougher_model_DecisionTree_depth = best_rougher_model_DecisionTree.get_depth()
print("Среднее sMAPE в результате кросс-валидации наилучшей модели дерева решений на обучающей черновой выборке:\n", 
      best_sMAPE_rougher_DecisionTree)
print("Глубина наилучшей модели дерева:", best_rougher_model_DecisionTree_depth)
```

У модели дерева решений при наилучшей глубине значение средней ошибки для финальной выборки ближе к 0, чем в линейной регрессии, тогда как в черновой получилось примерно то же самое.

#### Случайный лес

В лесу нужно перебрать два гиперпараметра. Сначала перебираем глубину на малом количестве деревьев, затем уже подбираем число оценщиков.

```python
best_model_RandomForest = None
best_rougher_model_RandomForest = None
best_sMAPE_RandomForest = 100
best_sMAPE_rougher_RandomForest = 100

for depth in range(1,21):    
    model_RandomForest = RandomForestRegressor(random_state=42, max_depth=depth, n_estimators=20)
    
    # final train set cross-validation sMAPE calculation
    sMAPE_RandomForest = cross_val_score(model_RandomForest, 
                                         features_train_scaled, 
                                         target_train, cv=5, 
                                         scoring=make_scorer(sMAPE, 
                                                             greater_is_better=False)).mean()
   
    
    # choice of best absolute values of sMAPE for final set
    if abs(sMAPE_RandomForest) < abs(best_sMAPE_RandomForest):
        best_model_RandomForest = model_RandomForest
        best_sMAPE_RandomForest = abs(sMAPE_RandomForest)        

    

for depth in range(1,21):    
    model_rougher_RandomForest = RandomForestRegressor(random_state=42, max_depth=depth, 
                                                       n_estimators=20)        
   
    # rougher train set cross-validation sMAPE calculation
    sMAPE_rougher_RandomForest = cross_val_score(model_rougher_RandomForest, 
                                         features_rougher_train_scaled, 
                                         target_rougher_train, cv=5, 
                                         scoring=make_scorer(sMAPE, 
                                                             greater_is_better=False)).mean()
    # choice of best absolute values of sMAPE for rougher set
    if abs(sMAPE_rougher_RandomForest) < abs(best_sMAPE_rougher_RandomForest):
        best_rougher_model_RandomForest = model_rougher_RandomForest
        best_sMAPE_rougher_RandomForest = abs(sMAPE_rougher_RandomForest)  
   
                
    
# чтобы получить глубину, обучаем лучшие модели, затем выводим метрику и гиперпараметр на экран       
best_model_RandomForest.fit(features_train_scaled, target_train)
best_model_RandomForest_depth = best_model_RandomForest.get_params().get('max_depth')        
print("Среднее sMAPE в результате кросс-валидации наилучшей модели случайного леса на обучающей финальной выборке:\n", 
      best_sMAPE_RandomForest)
print("Глубина наилучшей модели случайного леса:", best_model_RandomForest_depth)   

best_rougher_model_RandomForest.fit(features_rougher_train_scaled, target_rougher_train)
best_rougher_model_RandomForest_depth = best_rougher_model_RandomForest.get_params().get('max_depth')
print("Среднее sMAPE в результате кросс-валидации наилучшей модели случайного леса на обучающей черновой выборке:\n", 
      best_sMAPE_rougher_RandomForest)
print("Глубина наилучшей модели случайного леса:", best_rougher_model_RandomForest_depth)
```

Метрики чуточку лучше, чем у моделей дерева. Вероятно, в этом проекте победит лес, нужно только подобрать количество деревьев для него.

```python
best_model_RandomForest2 = None
best_rougher_model_RandomForest2 = None
best_sMAPE_RandomForest2 = 100
best_sMAPE_rougher_RandomForest2 = 100

# final set
for est in range(1,51):
    
    model_RandomForest2 = RandomForestRegressor(random_state=42, max_depth=2, n_estimators=est)
    
    # final train set cross-validation sMAPE calculation
    sMAPE_RandomForest2 = cross_val_score(model_RandomForest2, 
                                         features_train_scaled, 
                                         target_train, cv=5, 
                                         scoring=make_scorer(sMAPE, 
                                                             greater_is_better=False)).mean()
   
    # choice of best absolute values of sMAPE for final set
    if abs(sMAPE_RandomForest2) < abs(best_sMAPE_RandomForest2):
        best_model_RandomForest2 = model_RandomForest2
        best_sMAPE_RandomForest2 = abs(sMAPE_RandomForest2)        

        
        
#rougher set        
for est in range(1,51):
    
    model_rougher_RandomForest2 = RandomForestRegressor(random_state=42, max_depth=1, n_estimators=est)
    
        # rougher train set cross-validation sMAPE calculation
    sMAPE_rougher_RandomForest2 = cross_val_score(model_rougher_RandomForest2, 
                                         features_rougher_train_scaled, 
                                         target_rougher_train, cv=5, 
                                         scoring=make_scorer(sMAPE, 
                                                             greater_is_better=False)).mean()    

    # choice of best absolute values of sMAPE for rougher set
    if abs(sMAPE_rougher_RandomForest2) < abs(best_sMAPE_rougher_RandomForest2):
        best_rougher_model_RandomForest2 = model_rougher_RandomForest2
        best_sMAPE_rougher_RandomForest2 = abs(sMAPE_rougher_RandomForest2)  
        
        
# чтобы получить глубину, обучаем лучшие модели, затем выводим метрику и гиперпараметр на экран       
best_model_RandomForest2.fit(features_train_scaled, target_train)
best_model_RandomForest2_est = best_model_RandomForest2.get_params().get('n_estimators')        
print("Среднее sMAPE в результате кросс-валидации наилучшей модели случайного леса на обучающей финальной выборке:\n", 
      best_sMAPE_RandomForest2)
print("Количество оценщиков наилучшей модели случайного леса:", best_model_RandomForest2_est)   

best_rougher_model_RandomForest2.fit(features_rougher_train_scaled, target_rougher_train)
best_rougher_model_RandomForest2_est = best_rougher_model_RandomForest2.get_params().get('n_estimators')
print("Среднее sMAPE в результате кросс-валидации наилучшей модели случайного леса на обучающей черновой выборке:\n", 
      best_sMAPE_rougher_RandomForest2)
print("Количество оценщиков наилучшей модели случайного леса:", best_rougher_model_RandomForest2_est)
```

Ого. Отклонение для черновой выборки тут сильно выше. Но для финаьной ошибка получилась меньше, и это есть хорошо, ведь она составляет три четверти общей метрики. Посмотрим для этих же признаков на дамми-моделях.

```python
dummy_mean = DummyRegressor(strategy = 'mean')
dummy_median = DummyRegressor(strategy = 'median')

dummy_mean.fit(features_train_scaled, target_train)
predictions_dummy_mean = dummy_mean.predict(features_test_scaled)
sMAPE_dummy_mean = sMAPE(target_test, predictions_dummy_mean)
print('Dummy(mean) sMAPE for final test:', sMAPE_dummy_mean)

dummy_median.fit(features_train_scaled, target_train)
predictions_dummy_median = dummy_median.predict(features_test_scaled)
sMAPE_dummy_median = sMAPE(target_test, predictions_dummy_median)
print('Dummy(median) sMAPE for final test:', sMAPE_dummy_median)

dummy_mean.fit(features_rougher_train_scaled, target_rougher_train)
predictions_rougher_dummy_mean = dummy_mean.predict(features_rougher_test_scaled)
sMAPE_rougher_dummy_mean = sMAPE(target_rougher_test, predictions_rougher_dummy_mean)
print('Dummy(mean) sMAPE for rougher test:', sMAPE_rougher_dummy_mean)

dummy_median.fit(features_rougher_train_scaled, target_rougher_train)
predictions_rougher_dummy_median = dummy_median.predict(features_rougher_test_scaled)
sMAPE_rougher_dummy_median = sMAPE(target_rougher_test, predictions_rougher_dummy_median)
print('Dummy(median) sMAPE for rougher test:', sMAPE_rougher_dummy_median)
```

Финальная проходит под границами "медианных" и "средних" предсказаний, а вот черновая выборка сильно подвела. 

### Итоговые оценки качества предсказаний эффективности обогащения. Выводы

```python
print("Итоговая метрика качества для обучающей выборки:",
     (0.25*sMAPE_train + 0.75*sMAPE_rougher_train))
print("Итоговая метрика качества для тестовой выборки:",
     (0.25*sMAPE_test + 0.75*sMAPE_rougher_test))
```

Проверка подобранных моделей на тестовой выборке показала результаты, худшие, чем полученные для фиктивных моделей. Вероятно, на шаге отбора признаков было сделано что-то некорректно, либо не сделано вовсе. Автор проекта был бы рад разобраться, в чём же дело :)
