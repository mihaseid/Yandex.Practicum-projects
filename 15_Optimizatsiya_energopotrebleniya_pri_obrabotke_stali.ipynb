С помощью успешно подобранной модели машинного обучения мы сможем находить те сочетания физико-химических характеристик и те сочетания присадок, при которых будут достигаться минимально возможные температуры стали, а, следовательно, будет меньше затрат электроэнергии.

```python
# функция для выявления выбросов (за пределами 1,5 межквартильных размахов);
# строит боксплот, сохраняет выбросы и печатает их количество относительно всего датасета;
# на вход столбец датафейма
def outliers_qualifier(column):
    column = pd.DataFrame(column)
    boxplot = plt.boxplot(column.values)
    plt.ylabel(column.columns[0])
    plt.title('"Ящик с усами" для признака: '+column.columns[0])
    plt.show()
    outliers = list(boxplot["fliers"][0].get_data()[1])
    print("Статистических выбросов у признака", column.columns[0], " -", 
          len(outliers), "объектов из", len(column), '\n',
         "или {:.0%}".format((len(outliers)/len(column))) )
  
```

```python
# минифункция для удаления выбросов за пределами 1,5 межквартильных размахов:
def del_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1   
    filter = (df[column] >= Q1 - 1.5 * IQR) & (df[column] <= Q3 + 1.5 * IQR)
    df = df.loc[filter] 
    return df
```

После небольшого ликбеза по электродинамике удалось выяснить следующее:
1. Если в данных фигурируют активная и реактивная мощности, значит, речь идёт о переменном токе :)
2. Активная и реактивная мощности влияют на одну полную мощность, а именно в векторной сумме дают вектор полной мощности, но сами эти величины описывают разные явления: активная описывает необратимые превращения электроэнергии в другие виды энергии, а реактивная описывает энергию сопротивления, возвращаемую от его источников (всяческие катушки и пр.). Оба признака важны для характеризации нагрузок электросети, которые как раз и являются предметом изменений и исследований в работе заказчика.
3. Сумма мощностей всех фаз за цикл производства - вполне легитимная характеристика.
4. Реактивная мощность не может быть по модулю больше активной, но может быть отрицательной.

У нас имеются данные по обоим видам мощностей в зависимости от времени измерения и номера партии. Так как мы объединяем данные по партиям, будем считать суммарные мощности по партиям.

Итак, что мы сделаем:
- Удалим данные с датой и временем за ненадобностью
- Проверим дубликаты
- Поменяем названия столбцов на более удобные
- Порадуемся, что в этих данных нет пропусков
- Сгруппируем данные по партиям, где в качестве финального значения напротив номера партии будет сумма активной и реактивной мощностей в партии, соответственно
- Из статистического резюме видно, что есть, например, огромная отрицательная реактивная мощность. Посмотрим на разброс, на аномалии и удалим или заместим их

### Случайный лес.

Старый добрый случайный лес требует подбора гиперпараметров, будем подбирать количество деревьев-оценщиков и глубину этих деревьев. Начнём с глубины.

```python
best_model_RandomForest = None
best_model_RandomForest_depth = 0
best_MAE_RandomForest = 10000


# ищем оптимальную глубину дерева от 0 до 50
for depth in range(1,51): 
    # создаём, обучаем и предсказываем:
    model_RandomForest = RandomForestRegressor(random_state=42, max_depth=depth, n_estimators=50)
    model_RandomForest.fit(features_train_scaled, target_train)
    predictions_valid = model_RandomForest.predict(features_valid_scaled)
    # считаем среднюю абсолютную ошибку
    MAE_RandomForest = mean_absolute_error(target_valid, predictions_valid)
   
    
    # выбираем лучшую модель по метрике MAE
    if MAE_RandomForest < best_MAE_RandomForest:
        best_model_RandomForest = model_RandomForest
        best_MAE_RandomForest = MAE_RandomForest        
        best_model_RandomForest_depth = best_model_RandomForest.get_params().get('max_depth')  

print("MAE модели с наилучшей глубиной дерева:\n", 
      best_MAE_RandomForest)
print("Лучшая глубина в модели случайного леса:", best_model_RandomForest_depth) 
```

Пока всё идёт к тому, что до более сложных моделей можно будет и не добираться: метрика великолепна! Подберём количество деревьев с учётом глубины 8.

```python
best_model_RandomForest = None
best_model_RandomForest_esimators = 0
best_MAE_RandomForest = 10000


# ищем оптимальное количество оценщиков с глубиной 8, проверяя числа от 1 до 100
for estimators in range(1,101):    
    model_RandomForest = RandomForestRegressor(random_state=42, max_depth=8, n_estimators=estimators)
    model_RandomForest.fit(features_train_scaled, target_train)
    predictions_valid = model_RandomForest.predict(features_valid_scaled)
  
    MAE_RandomForest = mean_absolute_error(target_valid, predictions_valid)   
  
    if MAE_RandomForest < best_MAE_RandomForest:
        best_model_RandomForest = model_RandomForest
        best_MAE_RandomForest = MAE_RandomForest        
        best_model_RandomForest_estimators = best_model_RandomForest.get_params().get('n_estimators')  

print("MAE модели с наилучшей глубиной дерева:\n", 
      best_MAE_RandomForest)
print("Лучшее количество оценщиков в модели случайного леса:", best_model_RandomForest_estimators) 
```

Метрика стала ещё немножечко лучше! Но, чтобы признать победу, нужно протестировать на свежих данных нашу модель. 

```python
%%time
model_RandomForest = RandomForestRegressor(random_state=42, max_depth=8, n_estimators=78)
model_RandomForest.fit(features_train_scaled, target_train)
predictions_test = model_RandomForest.predict(features_test_scaled)
MAE_RandomForest = mean_absolute_error(target_test, predictions_test) 
print("MAE модели с наилучшими гиперпараметрами для тестовой выборки:\n", 
      MAE_RandomForest)
```

Метрика также хороша, как и для валидационной выборки по той простой причине, что неугодные по причинам выбросов или недостаточности целевого признака партии были удалены ещё до разделения выборок. Соответственно, в нашем случае успешное применение модели случайного леса невозможно без аналогичной предобработки новых тестовых выборок. Как уже объяснялось выше, решение удалить партии заранее было вызвано необходимостью суммирования значений мощностей: если бы удаление выбросов мощности происходило до объединения в партии, это привело бы к получению некорректных сумм в ряде случаев. Поэтому из двух зол - сокращение данных ради реалистичности или пренебрежение корректностью части данных ради увеличения данных - автор выбрал первое меньшее зло.

Так как в данном случае достигнут хороший результат качества обучения и предсказания, пробовать более затратные по времени модели бустинга и нейросети автор считает нецелесообразным.

### Подводные камни и камни преткновения.

В процессе работы вполне ожидаемо возникали всяческие "сюрпризы".

1. <b>Сюрприз 1. Исходные данные. </b> Данные разбросаны по 7 разным файлам, да ещё и имеют разные размеры. Имелись как временные ряды, так и распределения параметров по партиям сплавов стали. Так как временные данные присутствовали не везде, а там, где имелись - не совпадали в разных сетах, потому что обозначали время замеров разных вещей - стало понятно, что перед нами отнюдь не задача временных рядов (автор выдохнул, хотя тоже умеет с ними обращаться!). Единственное, что объединяло наши данные - номер партии, которых их все и объединил. В финальный датасет не вошли данных о времени добавления присадок, что оказалось правильным решением, для обучения имеющихся признаков оказалось достаточно.
2. <b>Сюрприз 2. "Температурные пропуски". </b> Датасет по температуре снабжал нас как целевым, так и обычным признаком для обучения. Целевой - последняя измеренная температура в партии, обычный - первая. С первыми проблем не возникло, они были во всех партиях. А вот последние отсутствовали в большом проценте данных. Предстояло решить, как заполнить эти пропуски, чтобы не случилось утечки целевого признака, и чтобы значения оказались правдоподобными. Единственным решением выходило усреднение температур по партиям, и скорее всего пришлось бы исключить первые температуры. Перед тем, как совершать какие-то телодвижения в этом направлении, нужно было проверить, где расположены эти наши пропуски. Оказалось, что все без исключения пустые ячейки располагались в партиях, где есть всего одно, первое измерение температуры, и использовать её для заполнения пропуска последней температуры невозможно. Это дало автору карт-бланш на удаление этих партий.
3. <b>Сюрприз 3. Удаление партий и перспективы тестовой выборки. </b> Одна из заповедей дата-учёного - не "причёсывать" тестовую выборку. Мы не знаем, на каких данных будет предсказывать целевой признак заказчик, поэтому тестовая выборка, которая есть здесь и сейчас, должна быть максимально реалистичной, и именно на ней мы проверяем качество обучения. Но бывают случаи, когда нужно пойти другим путём, чаще всего, когда тестовая выборка изначально в составе исходной, как и в данном проекте. Таким "причёсыванием" здесь служило удаление целых неугодных партий из датасетов, которое было применено в ряде признаков, в том числе и в целевом. Возможно, удаление отдельных пропусков/аномалий было бы некритичным, если бы впоследствии не пришлось бы, например, суммировать данные по партиям. Активная и реактивная мощности, в частности, были подвержены суммированию перед тем, как попасть в финальный датасет. И, удали мы какие-то из значений внутри партий, получившиеся суммы нельзя было бы назвать объективными и точными, чтобы на них обучать машину. 

    Поэтому было решено обработать всё ещё до разделения на выборки, но обязательно оговорить, что процесс работы с моделью будет включать в себя предварительную обработку выбросов в каждой новой тестовой выборке. В качестве предобработки необходимо использовать эту функцию для удаления выбросов в тестовой выборке (данные объединены по партиям!) перед применением в обучении:
    
   `# минифункция для удаления выбросов за пределами 1,5 межквартильных размахов:
def del_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1   
    filter = (df[column] >= Q1 - 1.5 * IQR) & (df[column] <= Q3 + 1.5 * IQR)
    df = df.loc[filter] 
    return df`
